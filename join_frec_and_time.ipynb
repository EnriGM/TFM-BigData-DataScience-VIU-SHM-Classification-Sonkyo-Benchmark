```python
import os
import numpy as np
```

```python
freq_features_folders = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\data\a8_features_freq"
time_features_folders = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\data\a8_features"
time_freq_features_folders = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\data\a8_features_freq_and_time"
```

```python
for case_ in os.listdir(time_features_folders):
    if "Case_" in case_:
        case_time = np.load(os.path.join(time_features_folders, case_))
        case_freq = case_.replace("_a8_features.npy", "_a8_freq_features.npy")
        case_freq = np.load(os.path.join(freq_features_folders, case_freq))
        case_data = np.hstack((case_time, case_freq))
        new_name = case_.replace("_a8_features.npy", "_a8_freq_time_features.npy")
        with open(os.path.join(time_freq_features_folders, new_name), 'wb') as f:
            np.save(f, case_data)
```

```python
def prepare_features_x_y(file):
    
     label = file.split("_")[1]
     temperature = int(file.split("(")[1].split(")")[0])
     
     data = np.load(os.path.join(FEATURES_FOLDER, file))
     rows = data.shape[0]
     temperature_arr = np.array([[temperature]*rows]).T
     x = np.hstack((data, temperature_arr))
     y = np.array([label]*rows)
     
     return x, y

def get_all_features_x_y():
    x = None
    y = None
    for file in os.listdir(FEATURES_FOLDER):
        if "Case_" in file:
            x_case, y_case = prepare_features_x_y(file)
            if x is None:
                x = x_case
                y = y_case
            else:
                x = np.vstack((x, x_case))
                y = np.hstack((y, y_case))
                
    return x, y
```

```python
FEATURES_FOLDER = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\data\a8_features_freq_and_time"

x, y = get_all_features_x_y()

with open(r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\res\data_for_ml_all_std_force_quality_freq_time\x_for_ml.npy", 'wb') as f:
    np.save(f, x)
    
with open(r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\res\data_for_ml_all_std_force_quality_freq_time\y_for_ml.npy", 'wb') as f:
    np.save(f, y)
```
