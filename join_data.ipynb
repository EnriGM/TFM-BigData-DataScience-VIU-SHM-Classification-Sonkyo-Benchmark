```python
import os
import numpy as np
ORIGINFOLDER = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\data\a8_npy_freq"

def prepare_x_y(file):
    
     label = file.split("_")[1]
     temperature = int(file.split("(")[1].split(")")[0])
     
     data = np.load(os.path.join(ORIGINFOLDER, file))
     rows = data.shape[0]
     temperature_arr = np.array([[temperature]*rows]).T
     x = np.hstack((data, temperature_arr))
     y = np.array([label]*rows)
     
     return x, y

def get_all_x_y():
    x = None
    y = None
    for file in os.listdir(ORIGINFOLDER):
        if "Case_" in file:
            x_case, y_case = prepare_x_y(file)
            if x is None:
                x = x_case
                y = y_case
            else:
                x = np.vstack((x, x_case))
                y = np.hstack((y, y_case))
                
    return x, y
```

```python
x, y = get_all_x_y()

with open("x_for_ltsm.npy", 'wb') as f:
    np.save(f, x)
    
with open("y_for_ltsm.npy", 'wb') as f:
    np.save(f, y)
```
