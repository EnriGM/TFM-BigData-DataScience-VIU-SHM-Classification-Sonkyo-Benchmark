```python
import pywt
import numpy as np
import os
from datetime import datetime
from scipy import sparse
```

```python
folder = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\data\a8_npy"
img_folder = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\cwt\cwt_no_comon_scale_matrix"

for file in os.listdir(folder):
    if "Case_" in file:
        print(datetime.now())
        print(file)
        data = np.load(os.path.join(folder, file))
        for count, signal in enumerate(data):
            dt = 1/len(signal)
            fs = len(signal)
            frequencies = np.array([400, 1]) / fs # the recording is 1666 hrz and the shaker uses 400 hrz so for both reasons is a good compromise
            scale = pywt.frequency2scale('morl', frequencies)
            scales = np.geomspace(scale[0], scale[-1], num=200)
            cwtmatr, freqs = pywt.cwt(signal, scales, "morl")
            cwtmatr = np.abs(cwtmatr[:-1, :-1]) # absolute take absolute value of complex result
            time = np.linspace(0, 1, len(signal))
            img_name = file.split(".")[0] + str(count).zfill(4) + ".npy"
            dest_p = os.path.join(img_folder, img_name)
            cwtmatr = cwtmatr.astype("float32")
            cwtmatr[cwtmatr < 0.5] = 0
            cwtmatr = sparse.bsr_matrix(cwtmatr)
            sparse.save_npz(dest_p, cwtmatr)
```

```python
import pandas as pd
import os
import re

img_folder = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\cwt\cwt_no_comon_scale_matrix"
index = list(range(len(os.listdir(img_folder))))

df = pd.DataFrame(index=index, columns=["filename", "case", "temperature"])

pattern = re.compile(r"\([^\)]*\)+")
for count, file in enumerate(os.listdir(img_folder)):
    df.loc[count, "filename"] = file
    df.loc[count,"case"] = file.split("_")[1]
    df.loc[count,"temperature"] = int(pattern.findall(file)[0].replace("(", "").replace(")", ""))

df.to_csv(r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\cwt\cwt_no_comon_scale_case_matrix_temp_info.csv")
```
